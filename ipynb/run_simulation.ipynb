# run_simulation

This notebook runs the main simulation routine and displays the output.

```python
%pylab inline
import pypvsim
import pandas as p
import datetime as dt
```

```python
# define battery type with python dictionary
lead_dict = {'type' : 'lead acid',
             'cost' : 0.14,
             'battery_efficiency_curve' : {'output_power':[0, 1000],
                                           'efficiency':[0.75,   0.75]},
             'DOD' : 0.5,
             'life' : 2
             }
```

```python
# call simulation routine
pypvsim.run_simulation(lead_dict, inverter_type='typical', load_type='day', plot=True, verbose=True)
```

```python
pypvsim.run_simulation(lead_dict, inverter_type='typical', load_type='night', plot=True, verbose=True)
```

```python
pypvsim.run_simulation(lead_dict, inverter_type='typical', load_type='continuous', plot=True, verbose=True)
```
