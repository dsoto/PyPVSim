# Solar Declination

This is an example of using the Solar class and the declination function from PyPVSim to plot yearly solar declination variation.

```python
%%capture
%pylab inline
import matplotlib.dates as mdates
import matplotlib.pyplot as plt

import pypvsim
import numpy as np
```

```python
solar = pypvsim.Solar(lat=0)
dates = np.arange('2013-01-01','2013-12-31', np.timedelta64(10,'D'), dtype='datetime64')
dec = [solar.declination(d.item()) for d in dates]
```

```python
fig, ax = plt.subplots()
# use the matplotlib date formatters
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.plot(dates.astype(object), 
        np.degrees(dec))
#fig.autofmt_xdate()
ax.set_title('Solar Declination')
ax.set_ylabel('Degrees North of equator')
ax.grid()
plt.show()
```
