# Sun-Path Diagrams

This notebook demonstrates the creation of sun-path diagrams using PyPVSim.  The angles are calculated in the Solar class and a convenience function in plots.py creates curves for different dates and hours of the day to create the curves.  

```python
%pylab inline
import pypvsim
import datetime as dt
import numpy as np
```

```python
# you can plot_sun_path as a pure convenience function
import matplotlib.pyplot as plt
pypvsim.plot_sun_path(lat=25)
```

```python
# you can also pass in an axes object to get other behavior like file saving
fig, ax = plt.subplots(2,1)
pypvsim.plot_sun_path(lat=25, ax=ax[0])
pypvsim.plot_sun_path(lat=40, ax=ax[1])
fig.show()
#fig.savefig('test.pdf')
```
